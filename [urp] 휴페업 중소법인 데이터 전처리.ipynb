```python
dat2 = pd.read_excel('휴폐업 중소법인 재무보유 (2).xlsx', sheet_name='휴폐업이력')
dat2.head()
```

```python
data = pd.merge(org_finan4, dat2[['BIZ_NO', 'STAT_OCR_DATE']], on='BIZ_NO')
```

```python
# 신생기업 여부 구분 (0:미분류, 1:신생X, 2:신생)

data['New'] = 0
data['New'][np.where(np.trunc((data['STAT_OCR_DATE'] - data['ESTB_DATE'])/10000) >= 7)[0]] = 1
data['New'][np.where(np.trunc((data['STAT_OCR_DATE'] - data['ESTB_DATE'])/10000) < 7)[0]] = 2
```

```python
# 불필요한 컬럼 제거
data.drop(columns=['ESTB_DATE', 'STAT_OCR_DATE'], inplace=True)
```

```python
# 결측치 처리 (직원수 : 동일 산업군 내 직원수 평균)

na_idx = np.where(data['EMP_CNT'].isna())[0]

for i in na_idx:
    ind = data.Industry[i]
    data.EMP_CNT[i] = data.EMP_CNT[data.Industry == ind].mean(numeric_only = True)

    if data.Industry[i] == np.nan:
        data.EMP_CNT[i] = data.EMP_CNT[np.where(data['Industry'].isna())[0]].mean(numeric_only = True)
```

```python
# 직원수 결측 존재 행 제거

data.dropna(subset=['EMP_CNT'], axis=0, inplace=True)
data.reset_index(inplace=True)
```

```python
# 회사 구분 결측 처리
data.Stock_OX[np.where(data.Stock_OX.isna())[0]] = '미분류'
```
