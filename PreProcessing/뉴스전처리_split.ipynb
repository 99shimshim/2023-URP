```python
from tqdm import tqdm
os.chdir('C:\\Users\\elcax\\OneDrive - 성균관대학교\\바탕 화면\\SKKU\\자격증 및 인턴\\개인 포트폴리오\\[URP] 휴폐업예측\\crawl_news')

dfs = [pd.read_csv(file) for file in tqdm(file_list[:4754])]
combined_df = pd.concat(dfs, axis=0)
```

```python
combined_df = combined_df.reset_index(drop=True)
combined_df
```

```python
comp_list = [file.replace('news_search_result_', '').replace('.csv', '') for file in file_list]
comp_list
```

```python
combined_df['contents'] = combined_df['contents'].str.strip("[]").replace('', np.nan)
combined_df['contents']
```

```python
# Escape special characters in company names
comp_list_escaped = [re.escape(comp) for comp in comp_list[:4754]]

# Create the regular expression pattern
pattern = '|'.join(comp_list_escaped)

# Filter the DataFrame based on company names
filtered_df = combined_df[combined_df['contents'].str.contains(pattern, regex=True, na=False)]
```

```python
filtered_df = filtered_df.reset_index(drop=True)
filtered_df
```

```python
filtered_df = filtered_df[filtered_df['company'].str.len() >= 3]
filtered_df = filtered_df.reset_index(drop=True)
filtered_df
```

```python
filtered_df.iloc[np.where(filtered_df.company == '21세기')[0]]
filtered_df = filtered_df[filtered_df['company'] != '21세기']
filtered_df = filtered_df[filtered_df['company'] != '메리트']
filtered_df = filtered_df[filtered_df['company'] != '과일장터']
filtered_df = filtered_df[filtered_df['company'] != '그린에너지']
filtered_df = filtered_df[filtered_df['company'] != '그린에너지개발']
filtered_df = filtered_df[filtered_df['company'] != '그린식품']
filtered_df = filtered_df[filtered_df['company'] != '그린시티']
filtered_df = filtered_df[filtered_df['company'] != '그린에코']
filtered_df = filtered_df[filtered_df['company'] != '글로벌']
filtered_df = filtered_df[filtered_df['company'] != '글로리아']
filtered_df.reset_index(drop=True)
filtered_df
```

```python
pattern = r'[\\t\\t\\t\n■]'  # 제거하고자 하는 특수 문자 패턴


filtered_df['contents'] = filtered_df['contents'].apply(lambda x: re.sub(pattern, '', x))
filtered_df['contents']
```

```python
# 문장 분리 함수
def split_sentences(row):
    sentences = row['contents'].split('.')
    return pd.DataFrame({'company': [row['company']] * len(sentences), 'contents': sentences})

```

```python
# 문장 분리 적용
new_df = pd.concat(filtered_df.apply(split_sentences, axis=1).tolist(), ignore_index=True)

# 결과 출력
print(new_df)
```
